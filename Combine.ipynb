## Splitting Data

```python
data_pivot = data.pivot_table(index=['Symbol', 'Date'] )
data_pivot
```

```python
data_ADANIPORTS = data_pivot.loc['ADANIPORTS']
data_ADANIPORTS.to_csv('data_ADANIPORTS.csv')
```

## Indexes

```python
#Getting Indexes
index = set()
for i in data['Symbol']:
    index.add(i)
print('No. of indices = ', len(index))
print(index)
```

```python
df_pivot = data.pivot_table(index=['Date'], columns='Symbol', values='Close')
df_pivot.index = pd.to_datetime(df_pivot.index)
df = df_pivot.sort_index()
df
```

```python
plt.figure(figsize=(30,15))
plt.plot(df)
plt.legend(df.columns)
plt.show()
```
